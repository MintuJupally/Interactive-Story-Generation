```python
import matplotlib.pyplot as plt
```

```python
def plot_loss():
    # Initialize lists for epoch and loss values
    loss_epochs = [1, 2, 3, 4, 5]
    loss_values = [2.1596, 2.0822, 2.02, 2.072500, 2.001000]
    eval_loss_epochs = [1, 2, 3, 4, 5]
    eval_loss_values = [2.152684, 2.132901, 2.137868, 2.213701, 2.226919]

    # Create the plot
    plt.figure(figsize=(10, 5))

    # Plot loss and eval_loss on the same graph with different line styles
    plt.plot(loss_epochs, loss_values, label='Training Loss', marker='o')
    plt.plot(eval_loss_epochs, eval_loss_values,
             label='Evaluation Loss', linestyle='--', marker='x')

    # Label the axes and the graph
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.title("Gemma-7b | Batch Size = 8 | Learning Rate = 2e-5 | (A100)")
    plt.legend()

    # Display the grid and the plot
    plt.grid(True)
    plt.show()
```

```python
plot_loss()
```
