```python
import matplotlib.pyplot as plt
import json
```

```python
def plot_loss(json_path, title):
    # Load your JSON data
    with open(json_path, 'r') as file:
        data = json.load(file)

    # Initialize lists for epoch and loss values
    loss_epochs = []
    loss_values = []
    eval_loss_epochs = []
    eval_loss_values = []

    # Extract values from JSON data
    for entry in data['log_history']:
        epoch = entry.get('epoch')
        if 'loss' in entry:
            loss_epochs.append(epoch)
            loss_values.append(entry['loss'])
        if 'eval_loss' in entry:
            eval_loss_epochs.append(epoch)
            eval_loss_values.append(entry['eval_loss'])

    # Create the plot
    plt.figure(figsize=(10, 5))

    # Plot loss and eval_loss on the same graph with different line styles
    plt.plot(loss_epochs, loss_values, label='Training Loss', marker='o')
    plt.plot(eval_loss_epochs, eval_loss_values, label='Evaluation Loss', linestyle='--', marker='x')

    # Label the axes and the graph
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.title(title)
    plt.legend()

    # Display the grid and the plot
    plt.grid(True)
    plt.show()
```

```python
plot_loss(json_path='./checkpoint-ep8-gpt2-med-b4-1e-5/trainer_state.json', title='GPT2-med | Batch Size: 4 | Learning Rate: 1e-5')
```

```python
# GPT2 small
# Batch Size: 4
# Epochs: 50


plot_loss(json_path='../finetuned/gpt2-small-bs4-ep20/trainer_state.json', title='GPT2-small | Batch Size: 4 | Learning Rate: 1e-5 (P100)')
```
